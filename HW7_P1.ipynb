```python
import numpy as np
import cv2

# Load an color image in grayscale
sky = cv2.imread('/home/ec2-user/sky.jpg',1)
bird = cv2.imread('/home/ec2-user/bird.jpg',1)
sky = cv2.resize(sky,(224,224))
bird = cv2.resize(bird,(224,224))
```

```python
merged = cv2.addWeighted(bird,0.7,sky,0.3,0)
#cv2.imwrite(merged)
subImage=merged-bird
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
plt.imshow(sky)
plt.title('original Sky')

plt.figure()
plt.imshow(bird)
plt.title('Original Bird')

plt.figure()
plt.imshow(merged)
plt.title('Merged')

plt.figure()
plt.imshow(subImage)
plt.title('Subracted Bird from Merged')
```
